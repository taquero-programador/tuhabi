```python
# version: 0.14

import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from urllib.parse import quote
import pandas as pd


driver = webdriver.Chrome()
driver.maximize_window()

for municipio, fraccionamientos in inmo.items():
    for fraccionamiento, modelos in fraccionamientos.items():
        for modelo, detalles in modelos.items():
            base_url = f'https://www.javer.com.mx/desarrollos/{fraccionamiento}/propiedad/{modelo}'
            url = base_url.format(
                fraccionamiento=quote(fraccionamiento, safe=''),
                modelo=quote(modelo, safe='')
            )
            driver.get(url)
            
            try:
                precio_lista = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.XPATH, "//p[contains(text(), 'Lista de precios')]"))
                )
                
                precio_lista.click()
                
                # sopa.find_all("h2", class_ = "woocommerce-loop-product__title")
                details = []
                soup = BeautifulSoup(driver.page_source, 'html.parser')
                direccion = soup.find('div', class_ = 'building-info-col').get_text()  # address
                precio = soup.find('p', class_ = 'property-price').get_text()  # price
                area_terreno = None #soup.find('p', class_ = 'property-price').get_text() # pxm
                area_construida = None # info
                dt = soup.find('div', class_ = 'bis-icon-row bir-1') # details
                for d in dt:
                    if d.text:
                        details.append(d.text[0:3])
                
                
                inmo[municipio][fraccionamiento][modelo]['address'] = direccion
                inmo[municipio][fraccionamiento][modelo]['price'] = precio.replace('$', '')
                inmo[municipio][fraccionamiento][modelo]['ground_area'] = area_terreno
                inmo[municipio][fraccionamiento][modelo]['area'] = area_construida
                inmo[municipio][fraccionamiento][modelo]['details'] = details
                
            except Exception as e:
                print(f"Error al procesar la página {url}: {str(e)}")

driver.quit()

def clean(args):
    return args.replace('-', ' ').title()


def address_clean(args):
    for char in args:
        pass

df = pd.DataFrame([(clean(municipio), clean(fraccionamiento), clean(modelo), detalles['address'],
                    detalles['price'], detalles['ground_area'], detalles['area'], *detalles['details']) 
                   for municipio, fraccionamientos in inmo.items() 
                   for fraccionamiento, modelos in fraccionamientos.items() 
                   for modelo, detalles in modelos.items()],
                  columns=['Municipio', 'Fraccionamiento', 'Model', 'Address', 'Price',
                           'Ground area', 'Area', 'bathrooms', 'Rooms', 'Parkings'])

df
```
