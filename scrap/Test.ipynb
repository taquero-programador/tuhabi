```python
data = {"Name": [], "Price": []}

for i in range(1,49):
    page = requests.get(f"{poke}page/{i}")
    sopa = BeautifulSoup(page.text, "html")
    pk = sopa.find_all("h2", class_ = "woocommerce-loop-product__title")
    data["Name"].extend([i.text.strip() for i in pk])
    pp = sopa.find_all("span", class_ = "woocommerce-Price-amount amount")
    data["Price"].extend([p.text.strip() for p in pp if p.text.strip() not in "£0.00"])

df2 = pd.DataFrame(data, index=range(1, len(data["Name"])+1))
df2.to_csv("pokemons.csv")
```
